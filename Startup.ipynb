# Separate x,y

```python
x=data.drop(['Profit'],axis=1)
```

```python
x
```

```python
y=data['Profit']
```

```python
y
```

# One hot encoding

```python
data=pd.get_dummies(x['State'])
```

```python
data
```

```python
x = x.drop('State',axis=1)
```

```python
x
```

```python
x = pd.concat([x,data],axis=1)
```

```python
x
```

# Splitting the dataset into the training set and test set

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor= LinearRegression()
```

```python
regressor.fit(xtrain,ytrain)
```

```python
xtest
```

```python
ytest
```

#  Predict the test set result

```python
pred=regressor.predict(xtest)
```

```python
pred
```

#  Calculate MAE

```python

from sklearn.metrics import mean_absolute_error
 
mae = mean_absolute_error(pred,ytest)
print("Mean Absolute Error is :" ,mae)
```

# MSE

```python
from sklearn.metrics import mean_squared_error
 
mse = mean_squared_error(pred, ytest)
print("Mean Squarred Error is :" ,mse*100)
```

# RMSE

```python
rmse = np.sqrt(mean_squared_error(pred, ytest))
print("Root Mean Squarred Error is :",rmse*100)
```

# MAPE

```python
from sklearn.metrics import mean_absolute_percentage_error
mape = mean_absolute_percentage_error(ytest, pred)
print("Mean Absolute Percentage Error",mape)
```

# R squared value

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(ytest,pred)
```

```python
score
```
